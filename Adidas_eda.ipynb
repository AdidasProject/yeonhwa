## product

### product별 total_sales

```python
product01 = pd.read_csv('/content/drive/MyDrive/project01_adidas/product_result01.csv')
product01
```

## retailer

### 시장 점유율

```python
retailer02 = pd.read_csv('/content/drive/MyDrive/project01_adidas/retailer02.csv')
retailer02
```

```python
retailer02 = retailer02.sort_values('retailer',ascending=False)
plt.title("Market Share by Retailer")
plt.pie(retailer02['market_share'],labels = retailer02['retailer'], autopct = '%0.1f%%')
plt.show()
```

1. 시장점유율
- retailer 6개
- West Gear, Foot Locker가 1,2 위를 차지하고 있음.
- 상위 그룹 3개와 하위 그룹 3개 간의 차이가 큼.
- 2020,2021년 합친 그래프도 2021년 매출이 2020년보다 커서 그 영향으로 인해, 2021년 그래프와 더 유사한 모습을 하고 있는 것으로 보임.

### 2021, 2022 시장점유율

```python
retailer05 = pd.read_csv('/content/drive/MyDrive/project01_adidas/retailer05.csv')
retailer05
```

```python
# 2020
retailer05_2020 = retailer05[retailer05['year']==2020]
retailer05_2020 = retailer05_2020.sort_values('retailer',ascending=False)

plt.title("Market Share 2020 by Retailer")
plt.pie(retailer05_2020['market_share'],labels = retailer05_2020['retailer'], autopct = '%0.1f%%')
plt.show()
```

```python
# 2021
retailer05_2021 = retailer05[retailer05['year']==2021]
retailer05_2021 = retailer05_2021.sort_values('retailer',ascending=False)

plt.title("Market Share 2021 by Retailer")
plt.pie(retailer05_2021['market_share'],labels = retailer05_2021['retailer'], autopct = '%0.1f%%')
plt.show()
```

2. 2020 → 2021 시장 점유율의 변화
- 2020년에 West Gear가 거의 50%를 차지하고 있었는데, 2021년에는 Foot Locker, Sports Direct, West Gear가 각각 20% 조금 넘게 차지하고 있음.
- 2020 → 2021년이 되면서 더 많은 소매점이 생겼으며, 독점 점유율이 낮아졌음을 확인할 수 있다.
